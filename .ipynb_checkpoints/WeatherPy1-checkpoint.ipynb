## WeatherPy

```python
import pandas as pd
import numpy as np
import requests
import scipy.stats as stats
from citipy import citipy
import matplotlib.pyplot as plt

api_key = 'fce8ae3f669e45fa2754271a1b8ea992'
```

## Generate Cities List

```python
longtitude = np.random.uniform(low = -180, high = 180, size= 2000)
latitude = np.random.uniform(low = -90 , high = 90, size=2000)

lat_lngs = zip(latitude, longtitude)
#list(lat_lngs)

cities = []
for lat_lng in lat_lngs:
    city= citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)
len(cities)
```

## Perfrom API Call

```python
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
index_c = 0
set_c = 1
```

```python
print('Retrieving City Data')
print('-----------------------------------------')


url = 'http://api.openweathermap.org/data/2.5/weather?'
units = 'imperial'
query = '{}appid={}&units={}&q='.format(url,api_key,units)

# For loop and create parameters
for i, city in enumerate(cities, start=1):
    try:
        response = requests.get(query+city).json()
        city_name.append(response['name'])
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        wind_speed.append(response['wind']['speed'])
        if index_c > 49:
            index_c = 0
            set_c += 1
        else:
            index_c += 1
            
        print('Processing Record {} of Set {}: {}'.format(index_c, set_c, city))
        
    except(KeyError, IndexError):
        print('City not found. Skipping....')
        
        
print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

```python
# Create dataframe CityWeather (cw) from list of data
cw = pd.DataFrame({
    'city': city_name, 'country':country, 'date': date,
    'max_temp': max_temp, 'cloudiness': cloudiness, 'humidity': humidity,
    'wind_speed': wind_speed, 'lat': lat, 'lng': lng
}).sort_values('city').reset_index().drop('index', axis=1)

print(cw.count())
cw.head()
```

```python
fig = plt.figure(figsize=(17,10))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.scatter(cw['lat'], cw['max_temp'], facecolor='blue', edgecolor='black')
ax1.set_title('Latitude vs. Max Temp (F)')
ax1.set_xlabel('Latitude')
ax1.set_ylabel('Max Temp')

ax2.scatter(cw['lat'], cw['humidity'], facecolor='red', edgecolor='black')
ax2.set_title('Latitude vs. Humidity (%)')
ax2.set_xlabel('Latitdue')
ax2.set_ylabel('Humidity (%)')

ax3.scatter(cw['lat'], cw['cloudiness'], facecolor='green', edgecolor='black')
ax3.set_title('Latitude vs. Cloudiness')
ax3.set_xlabel('Latitude')
ax3.set_ylabel('Cloudiness')

ax4.scatter(cw['lat'], cw['wind_speed'], facecolor='orange', edgecolor='black')
ax4.set_title('Latitude vs. Wind Speed (MPH)')
ax4.set_xlabel('Latitude')
ax4.set_ylabel('Wind Speed')

plt.tight_layout()
plt.show()
```

**Latitude vs. Max Temp (F):** As you get closer to the equator (lat of 0) temperatures rise.  They stay at their highest between -20 and 20 degrees latitude and start to drop off as the coordinates move more north or south.

**Latitude vs Humidity (%):** As the latitude gets closer to 0, many cities start to see a drop in humidity.  Although for most, humidity is around the same.  This may relate to cities around water.



```python
# Create northern and souther hemispheres
north = cw[cw.lat >= 0]
south = cw[cw.lat < 0]
```

```python
# Define linear regression function and create a plot
def LinReg_Scatter(x,y):
    print('The r^2 is : {}'.format(round(stats.pearsonr(x,y)[0],2)))
    slope, intercept, r_value, p_value, std_error = stats.linregress(x,y)
    regress_vals = x * slope + intercept
    line = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))
    plt.scatter(x,y)
    plt.plot(x, regress_vals, 'r-')
    print(line)
```

```python
LinReg_Scatter(north.lat, north.max_temp)

plt.title('Northern Hemisphere Linear Regression- Latitude vs. Max Temp')
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
plt.show()

#------------------------------------------

LinReg_Scatter(south.lat, south.max_temp)

plt.title('Southern Hemisphere Linear Regression- Latitude vs. Max Temp')
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
plt.show()
```

```python
LinReg_Scatter(north.lat, north.humidity)

plt.title('Northern Hemisphere Linear Regression- Latitude vs. Humidity (%)')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.show()

#------------------------------------------

LinReg_Scatter(south.lat, south.humidity)

plt.title('Southern Hemisphere Linear Regression- Latitude vs. Humidity (%)')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.show()
```

```python
LinReg_Scatter(north.lat, north.cloudiness)

plt.title('Northern Hemisphere Linear Regression- Latitude vs. Cloudiness')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness')
plt.show()

#------------------------------------------

LinReg_Scatter(south.lat, south.cloudiness)

plt.title('Southern Hemisphere Linear Regression- Latitude vs. Cloudiness')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness')
plt.show()
```

```python
LinReg_Scatter(north.lat, north.wind_speed)

plt.title('Northern Hemisphere Linear Regression- Latitude vs. Wind Speed')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed')
plt.show()

#------------------------------------------

LinReg_Scatter(south.lat, south.wind_speed)

plt.title('Southern Hemisphere Linear Regression- Latitude vs. Wind Speed')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed')
plt.show()
```
