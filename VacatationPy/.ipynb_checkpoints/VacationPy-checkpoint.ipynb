```python

fig = gmaps.figure(center=(35, -5), zoom_level=2)

location = cw[['lat','lng']]
weight = cw.humidity.astype(float)

heat_layer = gmaps.heatmap_layer(locations=location, weights=weight, 
                                 dissipating=False, max_intensity=100, point_radius=4)

fig.add_layer(heat_layer)
fig
```

```python
# Filter minimum and maximum temperature, rainfall and snowfall for ideal vacation
print('What is the maximum temperature you prefer?')
max_temp = int(input())

print('What is the minimum temperature you prefer?')
min_temp = int(input())

print('Do you high or low windspeed? (high/low)')
wind_speed = input()

print('Do you want any cloudiness? (heavy/light)')
input()
cloudiness = 'no'

if wind_speed == 'high':
    customer_search = cw[cw['wind_speed'] > 15]
    customer_search = customer_search[
    (customer_search['max_temp'] >= min_temp) 
    & (customer_search['max_temp'] <= max_temp)
    ].sort_values(by='city').reset_index().drop('index', axis=1)

else:
    customer_search = cw[cw['wind_speed'] <= 15]
    customer_search = customer_search[
    (customer_search['max_temp'] >= min_temp) 
    & (customer_search['max_temp'] <= max_temp)
    ].sort_values(by='city').reset_index().drop('index', axis=1)
    
if cloudiness == 'heavy':
    customer_search = cw[cw['cloudiness'] >= 50]
    cw = cw[
    (cw['max_temp'] >= min_temp) 
    & (cw['max_temp'] <= max_temp)
    ].sort_values(by='city').reset_index().drop('index', axis=1)
else:
    customer_search = cw[cw['cloudiness'] < 50]
    cw = cw[
    (cw['max_temp'] >= min_temp) 
    & (cw['max_temp'] <= max_temp)
    ].sort_values(by='city').reset_index().drop('index', axis=1)

customer_search
```
